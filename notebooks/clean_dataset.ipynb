```python
import pandas as pd

df = pd.read_csv("data/processed/rtc_master_dataset.csv")

print(df.shape)
df.head()
```

```python
print(df.isnull().sum())
```

```python
print("Duplicate rows:", df.duplicated().sum())
print("Duplicate text:", df["text"].duplicated().sum())
```

```python
df["text"] = (
    df["text"]
    .str.replace(r"[\n\r\t]+", " ", regex=True)
    .str.replace(r" {2,}", " ", regex=True)
    .str.strip()
)
```

```python
print(df["text"].iloc[0])
```

```python
df["report_name_clean"] = (
    df["report_name"]
    .str.replace("-", "_", regex=False)
    .str.strip()
)
```

```python
df[["report_name","report_name_clean"]].head()
```

```python
df["text_length"] = df["text"].str.len()
```

```python
df["word_count"] = df["text"].str.split().str.len()
```

```python
print(df.shape)
print(df.columns.tolist())
```

```python
import os

os.makedirs("data/processed", exist_ok=True)
```

```python
df.to_csv(
    "data/processed/rtc_master_dataset_cleaned.csv",
    index=False
)

print("Saved Successfully!")
```

```python
check = pd.read_csv(
    "data/processed/rtc_master_dataset_cleaned.csv"
)

print(check.shape)
print(check.columns)
```

```python
summary = df.groupby("report_name_clean").agg({
    "page_number":"count",
    "word_count":"mean",
    "text_length":"mean"
}).round(2)

summary
```

```python
summary.to_csv(
    "data/processed/report_summary.csv"
)
```
