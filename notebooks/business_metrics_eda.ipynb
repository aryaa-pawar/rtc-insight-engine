```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("data/processed/rtc_business_metrics.csv")

print("Shape:", df.shape)
df.head()
```

```python
print("Columns:")
print(df.columns)

print("\nYears:")
print(df["year"].unique())

print("\nCategories:")
print(df["category"].unique())
```

```python
#kpi count
year_counts = df.groupby("year").size()

print(year_counts)

year_counts.plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Number of KPIs by Year")
plt.ylabel("Count")
plt.show()
```

```python
'''
2025 contains the largest number of business KPIs,
indicating richer reporting and impact measurement.
'''
```

```python
category_counts = df.groupby("category").size()

print(category_counts)

category_counts.sort_values().plot(
    kind="barh",
    figsize=(10,6)
)

plt.title("KPI Count by Category")
plt.xlabel("Count")
plt.show()

##Community, Career and Recruiting categories contain the largest concentration of metrics.
```

```python
top_metrics = df.sort_values(
    "metric_value",
    ascending=False
).head(15)

top_metrics[[
    "metric_name",
    "metric_value"
]]
```

```python
top_metrics = df.sort_values(
    "metric_value",
    ascending=False
).head(10)

plt.figure(figsize=(12,6))

plt.barh(
    top_metrics["metric_name"],
    top_metrics["metric_value"]
)

plt.title("Top 10 Highest RTC Metrics")
plt.xlabel("Metric Value")

plt.show()
```
